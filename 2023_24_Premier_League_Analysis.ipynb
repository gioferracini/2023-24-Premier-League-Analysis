**2023/24 Premier League Season Overview**

The 2023/24 season marked the 32nd edition of the Premier League and the 125th season of top-flight football in England. Kicking off on August 11, 2023, and concluding on May 19, 2024, the campaign delivered historic moments and compelling narratives.

Manchester City successfully defended their title, becoming the first club in Premier League history to win four consecutive championships. At the other end of the table, all three promoted teamsLuton Town, Burnley, and Sheffield Unitedwere relegated, a rare occurrence not seen since the 1997/98 season.

# **Variables**

*   Player: The name of the player.
*   Nation: The player's nationality.
*   Pos: The player's position (e.g., forward, midfielder, defender).
*   Age: The player's age.
*   MP (Minutes Played): Total minutes played by the player.
*   Starts: Number of matches the player started.
*   Min (Minutes): Total minutes played by the player (this might be the same as MP).
*   90s (90s Played): The equivalent of 90-minute matches played by the player (e.g., 1.5 = 135 minutes).
*   Gls (Goals): Total number of goals scored by the player.
*   Ast (Assists): Total number of assists made by the player.
*   G+A (Goals + Assists): Total number of goals and assists combined.
*   G-PK (Goals - Penalty Kicks): Total number of goals scored excluding penalty kicks.
*   PK (Penalty Kicks): Number of penalty goals scored by the player.
*   PKatt (Penalty Kicks Attempted): Number of penalty kicks attempted by the player.
*   CrdY (Yellow Cards): Number of yellow cards received by the player.
*   CrdR (Red Cards): Number of red cards received by the player.
*   xG (Expected Goals): The expected number of goals from the player's shots.
*   npxG (Non-Penalty Expected Goals): Expected goals excluding penalties.
*   xAG (Expected Assists): The expected number of assists from the player's passes.
*   npxG+xAG (Non-Penalty xG + xAG): Total of non-penalty expected goals and expected assists.
*   PrgC (Progressive Carries): Number of times the player carried the ball forward.
*   PrgP (Progressive Passes): Number of passes made by the player that moved the ball forward.
*   PrgR (Progressive Runs): Number of times the player made runs forward with the ball.
*   Gls (Goals): (Repeated, already defined) Total number of goals scored.
*   Ast (Assists): (Repeated, already defined) Total number of assists made.
*   G+A (Goals + Assists): (Repeated, already defined) Total number of goals and assists combined.
*   G-PK (Goals - Penalty Kicks): (Repeated, already defined) Goals scored excluding penalty kicks.
*   G+A-PK (Goals + Assists - Penalty Kicks): Total goals and assists minus penalty goals.
*   xG (Expected Goals): (Repeated, already defined) Expected number of goals from the player's shots.
*   xAG (Expected Assists): (Repeated, already defined) Expected number of assists from the player's passes.
*   xG+xAG (Expected Goals + Expected Assists): Total expected goals and assists.
*   npxG (Non-Penalty Expected Goals): (Repeated, already defined) Expected goals excluding penalties.
*   npxG+xAG (Non-Penalty xG + Expected Assists): Total of non-penalty expected goals and expected assists.

# **Analysis by Clubs**

```python
pl_clubs = premierleague[['Team', 'Gls', 'Ast', 'G+A', 'xG', 'xAG', '90s', 'PrgC', 'PrgP', 'PrgR', 'CrdY', 'CrdR']]
pl_clubs = pl_clubs.groupby('Team').agg({
    'Gls': 'sum',
    'Ast': 'sum',
    'G+A': 'sum',
    'xG': 'sum',
    'xAG': 'sum',
    '90s': 'sum',
    'PrgC': 'sum',
    'PrgP': 'sum',
    'PrgR': 'sum',
    'CrdY': 'sum',
    'CrdR': 'sum'
}).reset_index()
pl_clubs.sort_values('G+A', ascending = False, inplace = True)
pl_clubs
```

```python
figure, ax = plt.subplots(figsize = (10, 6))
g = ax.barh(pl_clubs['Team'], pl_clubs['Gls'], color = 'r', label = 'Goals')
a = ax.barh(pl_clubs['Team'], pl_clubs['Ast'], left = pl_clubs['Gls'], color = 'cyan', label = 'Ast')
ax.bar_label(g, label_type='center', color = 'white')
ax.bar_label(a, label_type='center')
ax.set_title('Total of Goals + Assists by Clubs')
ax.legend()
plt.show()
```

By the end of the season, Manchester City stood out as the most effective attacking team, scoring 94 goals and providing 69 assists  the highest in both metrics.

On the other hand, Sheffield United had the fewest goals scored (31), while Burnley registered the fewest assists (17).

```python
figure, ax = plt.subplots(figsize = (10, 6))

for i, row in pl_clubs.iterrows():
  plt.text(row['xG'] + 0.3, row['Gls'], row['Team'], fontsize=10)

cmap = plt.cm.get_cmap('tab20c', len(pl_clubs))
ax.scatter(pl_clubs['xG'], pl_clubs['Gls'], s = pl_clubs['Gls'] * 5, c = range(len(pl_clubs)), cmap = cmap, alpha = 0.7, label = 'Goals')
ax.set_title('Goals vs xG')
ax.set_xlabel('xG')
ax.set_ylabel('Goals')
plt.show()
```

This scatter plot shows the relationship between actual goals scored (Gls) and expected goals (xG) for each Premier League team in the 2023/24 season. The expected goals (xG) metric evaluates the quality of chances created, giving insight into whether a team is finishing efficiently or underperforming.

Teams above the diagonal (where Goals > xG) outperformed expectations, likely due to clinical finishing or standout individual performances. Notably, Manchester City stands out with the highest overperformance, scoring over 90 goals from approximately 85 xG. Arsenal and Newcastle United also slightly exceeded expectations.

On the other hand, teams below the diagonal (where Goals < xG) underperformed their expected goals, indicating inefficiencies in converting chances. For example, Everton and Brighton created chances equivalent to around 5055 xG but finished with fewer goals, suggesting missed opportunities.

Sheffield United, positioned at the bottom left, had both the lowest xG and the fewest goals, reflecting their struggles in attack throughout the season.

This comparison helps reveal not just how many goals teams scored, but also how effectively they capitalized on the chances they created.

```python
figure, ax = plt.subplots(figsize = (10, 6))

for i, row in pl_clubs.iterrows():
  plt.text(row['xAG'] + 0.3, row['Ast'], row['Team'], fontsize=10)

cmap = plt.cm.get_cmap('tab20c', len(pl_clubs))
ax.scatter(pl_clubs['xAG'], pl_clubs['Ast'], s = pl_clubs['Ast'] * 5, c = range(len(pl_clubs)), cmap = cmap, alpha = 0.7, label = 'Assists')
ax.set_title('Assists vs xAG')
ax.set_xlabel('xAG')
ax.set_ylabel('Ast')
plt.show()
```

This scatter plot compares actual assists (Ast) to expected assists (xA) for each team in the 2023/24 Premier League season. The xA metric estimates how many assists a team should have based on the quality of their key passes and chances created, regardless of whether the final shot resulted in a goal.

Teams located above the diagonal line (Ast > xA) exceeded expectations, converting chances into assists more efficiently. Manchester City again leads the way, with the highest overperformance in this category  delivering around 65 assists from a ~70 xA base. Arsenal, Tottenham, and Liverpool also performed strongly, slightly outperforming their expected assist values.

Teams below the line underperformed their creative potential. Notably, Everton registered significantly fewer assists (17) compared to their xA (40), suggesting problems in turning good chances into actual goals. Similarly, Sheffield United struggled both in creating and converting opportunities, finishing at the bottom in both metrics.

This plot highlights which teams were the most efficient in turning creative play into tangible attacking output  and which teams struggled to capitalize on their build-up play.

```python
figure, ax = plt.subplots(figsize = (10, 6))
p = ax.barh(pl_clubs['Team'], pl_clubs['PrgTotal'], color = 'springgreen')
ax.set_title('Total of Progression by Club')
ax.set_xlabel('Total of Progressions')
ax.bar_label(p, label_type='edge')
plt.show()
```

This bar chart displays the total number of progression actions by each Premier League team in the 2023/24 season. The metric includes a combined count of progressive passes, carries, and runs  actions that move the ball significantly closer to the opponents goal and reflect a team's attacking intent and territorial dominance.

As expected, the teams with stronger offensive profiles dominate the top of the chart. Manchester City leads the league with 5,219 total progressions, closely followed by Tottenham Hotspur, Liverpool, and Arsenal  all of whom exceeded 5,000 progression actions. These teams maintained high possession rates and were consistently proactive in advancing the ball.

In contrast, teams at the bottom of the chart like Sheffield United, Everton, and Crystal Palace had significantly fewer progression actions. Sheffield United, in particular, recorded only 2,101 progressions  the lowest in the league  reinforcing their overall struggles in build-up play and chance creation.

This metric highlights how often each team advanced the ball into more dangerous areas, offering a broader view of their playing style and offensive activity beyond just goals and assists.

The data reveals that defenders dominate in both carrying and passing progressions. This is consistent with their tactical role, where they are responsible for advancing the ball forward, either by making passes to midfielders or wingers, thus initiating offensive transitions. In contrast, attackers exhibit superior dribble progressions, reflecting their primary function of driving the ball into the attacking third through individual runs. This distinction highlights the complementary roles within the team, with defenders focusing on ball progression through distribution and attackers exploiting space via direct carries.
